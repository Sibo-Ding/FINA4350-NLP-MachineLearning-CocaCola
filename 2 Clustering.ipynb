```python
import pandas as pd
import matplotlib.pyplot as plt

bow = pd.read_csv('bag_of_words.csv')
sales = pd.read_csv('sales.csv')

bow_1 = bow.iloc[:, 1:] # Exclude "quarter_statement"
```

## K-means clustering

```python
from sklearn.cluster import KMeans

for no_cluster in range(2, 9):
    # Fit k-means model
    k_clust = KMeans(n_clusters=no_cluster, random_state=23)\
        .fit_predict(bow_1)
    
    # Visualize sales belong to which cluster
    plt.figure()
    plt.scatter(x=k_clust, y=sales['growth'], c=k_clust)
    plt.xlabel(f'{no_cluster} Clusters (K-means)')
    plt.ylabel('Growth of Sales')
```

## Hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering

for no_cluster in range(2, 9):
    h_clust = AgglomerativeClustering(
        n_clusters=no_cluster, metric='euclidean', linkage='ward')\
        .fit_predict(bow_1)
    
    plt.figure()
    plt.scatter(x=h_clust, y=sales['growth'], c=h_clust)
    plt.xlabel(f'{no_cluster} Clusters (Hierarchical)')
    plt.ylabel('Growth of Sales')
```
